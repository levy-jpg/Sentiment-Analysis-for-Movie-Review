```python
#Levy Thiga Kariuki
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('IMDB Dataset.csv')

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['review'], df['sentiment'], test_size=0.2, random_state=42)

# Create a pipeline for text classification
text_clf = Pipeline([
    ('vect', CountVectorizer()),  # Tokenization and vectorization
    ('clf', DecisionTreeClassifier()),  # Decision Tree classifier
])

# Train the model
text_clf.fit(X_train, y_train)

# Predictions on the test set
y_pred = text_clf.predict(X_test)

# Evaluate the performance
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Display classification report
print(classification_report(y_test, y_pred))

# Plot loss curve (not applicable for Decision Trees)
# Decision Trees do not have a loss curve as they don't use iterative optimization like neural networks

# Optionally, save the trained model
# joblib.dump(text_clf, 'decision_tree_sentiment_analysis_model.joblib')
```
